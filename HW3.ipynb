```python
def extract_label(token):
    fnd = re.findall('{(.*)}', token)
    return 0 if len(fnd) == 0 else (1 if fnd[0] == 'ORG' else 2)

def parse_sentence(text):
    tokens = re.findall('[\\w\-{}]+', text)
    labels = list(map(extract_label, tokens))
    text_without_tags = re.sub('{.*?}', '', text)
    seq = list(filter(lambda x: len(x) > 0, re.sub('[^\\w\-]+', ' ', ''.join(mystem.lemmatize(text_without_tags))).split(' ')))
    return seq, labels
```

```python
freq = {}
sents = []
labels = []
with open('train_sentences_enhanced.txt') as input:
    for sentence in tqdm(input.readlines()):
        seq, label = parse_sentence(sentence)
        if len(seq) != len(label):
            continue
        sents.append(seq)
        labels.append(label)
        for word in seq:
            if not word in freq:
                freq[word] = 0
            freq[word] += 1

words = list(freq.keys())
words.sort(key=(lambda x: freq[x]))

word_ind = {}
for i, word in enumerate(words):
    word_ind[word] = i + 1

train_seq, test_seq, train_label, test_label = train_test_split(sents, labels, test_size=0.2)
```

```python
def J(output, label):
    return -torch.sum(torch.log(output[torch.arange(len(label)).long(), label.long()]))
```

```python
class FNN(nn.Module):
    def __init__(self, dict_size, n):
        super(FNN, self).__init__()
        self.n = n
        self.embed = nn.Embedding(dict_size, 200)
        layers = [
            nn.Linear(200 * n, 40),
            nn.Sigmoid(),
            nn.Dropout(0.5),
            nn.Linear(40, 40),
            nn.Sigmoid(),
            nn.Dropout(0.5),
            nn.Linear(40, 3),
            nn.Softmax(dim=-1)
        ]
        self.model = nn.Sequential(*layers)

    def forward(self, x):
        t_pad = F.pad(x, pad=(self.n // 2, self.n // 2))
        #print(t_pad.dtype)
        embeds = self.embed(t_pad)
        final_embeds = embeds[torch.arange(len(x))[:, None] + torch.arange(self.n)]
        return self.model.forward(final_embeds.reshape(-1, 200 * self.n))
```

```python
model = FNN(len(words) + 1, 7)
adam = optim.Adam(model.parameters())
num_epochs = 150

train_losses = []
test_losses = []
for i in tqdm(range(num_epochs)):
    train_loss = torch.tensor(0.0)
    train_loss.requires_grad_()
    train_size = 0
    for sent, label in zip(train_seq, train_label):
        inds = torch.tensor(list(map(lambda x: word_ind[x], sent))).long()
        output = model(inds)
        #print(label)
        #print(output)
        #print(torch.tensor(label))
        loss = J(output, torch.tensor(label))
        train_loss = train_loss + loss
        #loss.backward()
        train_size += len(sents)
    train_loss = train_loss / torch.tensor(train_size).float()
    train_loss.backward()
    train_losses.append(train_loss)
    
    test_loss = 0
    test_size = 0
    for sent, label in zip(test_seq, test_label):
        inds = torch.tensor(list(map(lambda x: word_ind[x], sent)))
        output = model(inds)
        loss = J(output, torch.tensor(label))
        test_loss += float(loss.float())
        test_size += len(sents)
    test_losses.append(test_loss / test_size)
```

```python
plt.plot(range(num_epochs), train_losses)
plt.plot(range(num_epochs), test_losses)
plt.legend(['train', 'test'])
```

```python
def tokenize(text):
    lens = list(map(len, re.findall('[\w\-]+', text)))
    offsets = list(map(len, re.split('[\w\-]+', text)))
    text = ''.join(mystem.lemmatize(text))
    tokens = re.findall('[\w\-]+', text)
    ans = []
    fr = 0
    for token, ln, off in zip(tokens, lens, offsets[:-1]):
        fr += off
        ans.append((token, fr, ln))
        fr += ln
    return ans
```

```python
with open('dataset_40163_1.txt', 'r') as inf:
    with open('result.txt', 'w') as ouf:
        for text in inf.readlines():
            tok = tokenize(text)
            inds = torch.tensor(list(map(lambda x: 0 if (x[0] not in word_ind) else word_ind[x[0]], tok))).long()
            output = model(inds).argmax(dim=-1)
            ans = ''
            for ((lm, fr, ln), la) in zip(tok, output):
                if la == 1:
                    ans += f'{fr} {ln} ORG '
                elif la == 2:
                    ans += f'{fr} {ln} PERSON '
            ans += 'EOL\n'
            ouf.write(ans)
```

```python
import catboost
from catboost import CatBoostClassifier, Pool
import pandas as pd
import numpy as np
```

```python
def get_pos(token_info):
    return 
    
def extract_features(text):
    n = 7
    cat_features = {2, 3, 4, 5, 6, 7, 8, 10}
    
    ans = pd.DataFrame()
    names = ['variants', 'len', 'upper', 'lower', 'pos', 'gender', 'naming', 'alive', 'number', 'wt', 'qual']
    is_str = [False, False, False, False, True, True, True, True, True, False, True]
    n_features = len(names)
    features = [[] for _ in names]
    
    
    labels = []
    fr = []
    
    next_is_tag = False
    cur_label = 0
    
    cur_fr = 0
    new_fr = 0
    for token_info in mystem.analyze(text):
        cur_fr = new_fr
        new_fr += len(token_info['text'])
        if 'analysis' not in token_info:
            if token_info['text'] == '{':
                next_is_tag = True
            continue
        if next_is_tag:
            cur_label = 1 if token_info['text'] == 'ORG' else 2
            next_is_tag = False
            continue
        fr.append((cur_fr, new_fr - cur_fr))
        labels.append(cur_label)
        cur_label = 0
        features[0].append(len(token_info['text']))
        features[1].append(len(token_info['analysis']))
        features[2].append(token_info['text'].isupper())
        features[3].append(token_info['text'].islower())
        features[4].append(re.findall('[A-Za-z]+', token_info['analysis'][0]['gr'])[0] \
                           if len(token_info['analysis']) > 0 else 'None')
        genders = re.findall('(муж|жен|сред)', token_info['analysis'][0]['gr']) \
            if len(token_info['analysis']) > 0 else []
        features[5].append(genders[0] if len(genders) > 0 else 'None')
        naming = re.findall('(имя|фам)', token_info['analysis'][0]['gr']) \
            if len(token_info['analysis']) > 0 else []
        features[6].append(naming[0] if len(naming) > 0 else 'None')
        alive = re.findall('(од|неод)', token_info['analysis'][0]['gr']) \
            if len(token_info['analysis']) > 0 else []
        features[7].append(alive[0] if len(alive) > 0 else 'None')
        number = re.findall('(ед|мн)', token_info['analysis'][0]['gr']) \
            if len(token_info['analysis']) > 0 else []
        features[8].append(number[0] if len(number) > 0 else 'None')
        features[9].append(token_info['analysis'][0]['wt'] if len(token_info['analysis']) > 0 else 0)
        features[10].append(token_info['analysis'][0]['qual'] if len(token_info['analysis']) > 0 and 'qual' in token_info['analysis'][0] else 'None')
        
    
    result_features = []
    result_names = []
    result_cat_features =  np.concatenate(n * np.array(list(cat_features))[:,None] + np.arange(n))
    for i in range(n_features):
        #print(features[i])
        if is_str[i]:
            tmp = np.array(['None'] * (n // 2) + features[i] + ['None'] * (n // 2))
        else:
            tmp = np.array([-1] * (n // 2) + features[i] + [-1] * (n // 2))
        feats = tmp[np.arange(n)[:, None] + np.arange(len(features[i]))]
        result_features.extend(feats)
    
    result_names = []
    for i in range(n_features):
        for j in range(n):
            result_names.append(names[i] + str(j))
        
    
    labels.append(cur_label)
    labels = labels[1:]
    
    #ans['variants'] = variants
    #ans['upper'] = upper
    #ans['lower'] = lower
    ans = pd.DataFrame(dict(zip(result_names, result_features)))
    
    return ans, labels, fr, list(result_cat_features)
```

```python
labels = []
features = pd.DataFrame()
cat_features = []
with open('train_sentences_enhanced.txt') as input:
    for text in tqdm(input.readlines()):
        feature, label, _, cat_features = extract_features(text)
        labels.extend(label)
        features = pd.concat([features, feature])
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)
```

```python
model = CatBoostClassifier(
    iterations=500,
    learning_rate=0.2,
    bootstrap_type='MVS',
    custom_metric='F1',
    loss_function='MultiClass'
)
model.fit(
    X_train, y_train,
    cat_features=cat_features,
    eval_set=(X_test, y_test),
    verbose=False,
    plot=True
)
#0.968, 0.574, 0.921 MultiClassOneVsAll MVS-bootstrap n=7
#0.969, 0.581, 0.919 MultiClass MVS-bootstrap n=7
#0.971, 0.6, 0.932 MultiClass MVS-bootstrap n=7 +2 extra features
#0.971, 0.598, 0.932 MultiClass n=7 +2 extra features
#0.973, 0.6, 0.933
```
